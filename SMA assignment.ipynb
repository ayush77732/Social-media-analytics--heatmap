```python
#importing necessary library
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from textblob  import TextBlob
import folium
from folium.plugins import HeatMap
```

```python
#Reading Json file
file_path="/content/dutch_tweets_chunk0.json"
df2=pd.read_json(file_path)
df2.tail()
```

```python
#Poting map using folium
df2=df2[(df2["latitude"].between(-90,90))&(df2["longitude"].between(-180,180))]
print(f"total records loaded:{len(df2)}")
map_center=[df2["latitude"].mean(),df2["longitude"].mean()]
heatmap_map=folium.Map(location=map_center,zoom_start=4)
heat_data=df2[["latitude","longitude"]].values.tolist()
HeatMap(heat_data,radius=10,blur=15,max_zoom=8).add_to(heatmap_map)
```

```python
#Generating and storing heatmap
output_file="desc_heatmap_full.html"
heatmap_map.save(output_file)
print(f"heatmap is save to:{output_file}")
```

```python
#Adding sentiment column to DataFrame and storing it into a csv file
def get_sentiment(text):
  analysis=TextBlob(str(text))
  polarity=analysis.sentiment.polarity
  if polarity>0:
    return 'Positive'
  elif polarity<0:
    return 'Negative'
  else:
    return 'Neutral'
df2['sentiment']=df2['desc_translation'].apply(get_sentiment)
df2.head()
output_file="data.csv"
df2.to_csv(output_file)
print(f"data is save to:{output_file}")
```
